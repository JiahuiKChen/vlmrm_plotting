### Plotting

```python
# convert step col to global step 
clip_reward_csv["Step"] = clip_step_csv[clip_step_csv.columns[1]]
gt_reward_csv["Step"] = gt_step_csv[gt_step_csv.columns[1]]

# dropping min/max cols 
clip_reward_csv.drop([col for col in clip_reward_csv.columns if 'MAX' in col],axis=1,inplace=True)
clip_reward_csv.drop([col for col in clip_reward_csv.columns if 'MIN' in col],axis=1,inplace=True)

# renaming SIGLIP runs to have: reward func, multi or single prompt
def rename_clip(name):
    # which runs are multi-prompt and which are single prompt
    mp = {'193951', '193614', '193423', '193211', '192935', '192417'}
    mp_neg = {'160730', '194149'}

    if "Step" in name:
        return name
        
    # getting reward func
    if "cosine" in name:
        r = "cosine"
    elif "contrastive" in name:
        r = "contrastive"
    elif "goal_baseline" in name:
        r = "goal_baseline_reg"

    id = name.split('_')[4 if "goal_baseline" not in name else 6].split()[0]
    # print(f"{id} \t {name}")
    if id in mp:
        mp = "multi-prompt"
    elif id in mp_neg:
        mp = "multi-prompt-negative"
    else:
        mp = "single-prompt"
    return f"{r}_{mp}"

clip_reward_csv = clip_reward_csv.rename(columns=rename_clip)

# clip_reward_csv.columns
```

```python
def moving_window_average(data, window_size):
    """Smooth data with moving window average."""
    if window_size <= 1:
        return data
    window = np.ones(int(window_size)) / float(window_size)
    return np.convolve(data, window, 'same')
```

```python
# TODO: add ground truth reward

clip_reward_csv = clip_reward_csv.interpolate(method="linear")
smooth_factor = 1
# plot a line for each reward function and prompt/multi-prompt combination
rewards = ["cosine", "contrastive", "goal_baseline_reg"]
# "multi-prompt-negative" for contrastive only
prompt_type = ["multi-prompt", "single-prompt"]
plt.figure(figsize=(10, 10))

for r in rewards:
    for p in prompt_type:
        name = f"{r}_{p}"

        # calculate average, std dev (2 runs with same name correspond to same seeds)
        means = clip_reward_csv[[name]].mean(axis=1)
        stds = clip_reward_csv[[name]].std(axis=1)
        # Apply smoothing
        smoothed_means = moving_window_average(means, smooth_factor)

        if r == "cosine":
            x = clip_step_csv['CartPole_siglip-base-patch16-224k_cosine_20240425_062033 - global_step']
        elif r == "contrastive":
            x = clip_step_csv['CartPole_siglip-base-patch16-224k_contrastive_20240426_194149 - global_step']
        elif r == "goal_baseline_reg":
            x = clip_step_csv['CartPole_siglip-base-patch16-224k_goal_baseline_reg_20240425_053932 - global_step']
        plt.plot(x, smoothed_means, label=name)
        plt.fill_between(x, smoothed_means-stds, smoothed_means+stds, alpha=0.2)

    if r == "contrastive":
        name = f"{r}_multi-prompt-negative"

        # calculate average, std dev (2 runs with same name correspond to same seeds)
        means = clip_reward_csv[[name]].mean(axis=1)
        stds = clip_reward_csv[[name]].std(axis=1)
        # Apply smoothing
        smoothed_means = moving_window_average(means, smooth_factor)

        plt.plot(clip_step_csv['CartPole_siglip-base-patch16-224k_contrastive_20240426_194149 - global_step'], smoothed_means, label=name)
        plt.fill_between(x, smoothed_means-stds, smoothed_means+stds, alpha=0.2)

    # ground truth reward
    env_reward = gt_reward_csv["CartPole_env_reward_20240422_232316 - rollout/ep_rew_mean"]
    plt.plot(gt_reward_csv["Step"], env_reward, label="Enviornment Reward")

# Labeling the plot
plt.xlabel('Step')
plt.ylabel('Score')
plt.title(f'Multiprompt vs Single Prompt Reward')
plt.legend(prop={'size': 10}, loc=(1.04, 0))
plt.show()
```
